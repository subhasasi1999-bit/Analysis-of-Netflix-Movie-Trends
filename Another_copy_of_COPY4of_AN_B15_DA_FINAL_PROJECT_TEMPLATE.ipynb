# **HISTOGRAM**

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(8,5))
plt.hist(sb['Popularity'], bins=20, color='skyblue', edgecolor='black')
plt.title('Distribution of Popularity')
plt.xlabel('Popularity')
plt.ylabel('Frequency')
plt.show()
```

# **Explanation**
1. This analysis helps identify whether most movies have low, medium, or high popularity.
2. If only a few movies are highly popular, the company can study their characteristics to create similar successful content.
3. It also helps in planning promotional strategies for less popular movies
# **Interpretation**

1. If most bars are on the left, most movies have low popularity.
2. A few bars on the right indicate highly popular movies.

# **COUNTPLOT **

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(8,5))
sns.countplot(x='Popularity', data=sb,color='skyblue' )
plt.title('Movies by Popularity')
plt.show()
```

# **Explanation**
1. The box plot identifies extremely popular movies (outliers).
2. These movies can be prioritized for recommendations, sequels, or special promotions.
3. Outliers may represent blockbuster movies that significantly contribute to user engagement.

# Interpretation

1. English may have the highest number of movies.
2. Other languages appear less frequently.

# **BOX PLOT**

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(14,9))
sns.boxplot(x= 'Vote_Average', y='Popularity', data=sb)
plt.title('Vote_Average by Popularity')
plt.xlabel('Vote_Average')
plt.ylabel('Popularity')
plt.grid(True)
plt.show()
```

# **Explanation**
1. Identifies movies that received exceptionally high numbers of votes.
2. Such movies are valuable assets for the platform because they often attract more viewers and improve platform credibility.

# Interpretation

Many movies may have low vote counts.
A few blockbuster movies have extremely high vote counts, appearing as outliers.

# **1. Popularity vs Vote Count (Scatter Plot)**
## **Objective:** Examine whether more popular movies receive more votes.

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(15,10))
sns.scatterplot(x='Vote_Count', y='Vote_Average', data=sb)

plt.title('Vote_Count vs Vote_Average')
plt.xlabel('Vote_Count')
plt.ylabel('Vote_Average')
plt.show()
```

# **Analysis**
1. Determines whether popular movies also receive more audience votes.
2. A strong positive relationship indicates that popular movies encourage greater audience engagement.
3. Helps identify blockbuster content.

# **Average Popularity by Original Language (Bar Plot)**

```python
plt.figure(figsize=(12,5))

language_popularity = sb.groupby('Original_Language')['Popularity'].mean().sort_values(ascending=False)

sns.barplot(
    x=language_popularity.index,
    y=language_popularity.values,
    palette='coolwarm'
)

plt.title('Average Popularity by Original Language')
plt.xlabel('Average Popularity')
plt.ylabel('Language')
plt.show()
```

# **Analysis**
1. Shows which language-based movies attract the highest popularity.
2. Supports decisions regarding regional content production and localization.

# **Correlation Heatmap**

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(9,7))

# Convert 'Vote_Count' and 'Vote_Average' to numeric, coercing errors
sb['Vote_Count'] = pd.to_numeric(sb['Vote_Count'], errors='coerce')
sb['Vote_Average'] = pd.to_numeric(sb['Vote_Average'], errors='coerce')

correlation = sb[['Popularity','Vote_Count','Vote_Average']].corr()

sns.heatmap(
    correlation,
    annot=True,
    cmap='coolwarm',
    linewidths=0.5
)

plt.title('Correlation Heatmap')
plt.show()
```

# **Analysis**
1. Summarizes relationships among popularity, vote count, and ratings.
2. Helps management quickly identify which performance metrics influence each other.
3. Useful for strategic planning and predictive analytics.

# **Pair Plot (Relationship among Numerical Variables**)
# **Variables: Popularity, Vote_Count, Vote_Average**

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.pairplot(
    sb[['Popularity', 'Vote_Count', 'Vote_Average']],
    diag_kind='hist'
)

plt.show()
```

# **Analysis**
1. Provides a complete overview of relationships among popularity, vote count, and ratings.
2. Helps identify trends, clusters, and potential correlations.
3. Useful during exploratory analysis before building predictive models.

#Bubble Plot (Advanced Plot)
## **Variables:** Popularity, Vote_Average, Vote_Count

```python
plt.figure(figsize=(10,6))

plt.scatter(
    sb['Popularity'],
    sb['Vote_Average'],
    s=sb['Vote_Count']/20,
    alpha=0.6
)

plt.xlabel('Popularity')
plt.ylabel('Vote Average')
plt.title('Popularity vs Vote Average (Bubble Size = Vote Count)')
plt.show()
```

# **Analysis**
Displays three variables simultaneously:
X-axis: Popularity
Y-axis: Vote Average
Bubble Size: Vote Count
Helps identify blockbuster movies that are popular, highly rated, and widely viewed.
Supports premium content recommendations.
